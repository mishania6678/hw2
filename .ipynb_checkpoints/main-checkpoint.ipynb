## Задание 2

##### С помощью функции linspace из библиотеки Numpy создайте массив t из 51 числа от 0 до 10 включительно.


```python
t = np.linspace(0, 10, 51)
t
```

##### Создайте массив Numpy под названием f, содержащий косинусы элементов массива t.

```python
f = np.cos(t)
f
```

##### Постройте линейную диаграмму, используя массив t для координат по горизонтали,а массив f - для координат по вертикали. Линия графика должна быть зеленого цвета.


```python
plt.plot(t, f, color='g')
plt.show()
```

##### Выведите название диаграммы - 'График f(t)'. Также добавьте названия для горизонтальной оси - 'Значения t' и для вертикальной - 'Значения f'.


```python
plt.plot(t, f, color='g')
plt.title('График f(t)')
plt.xlabel('Значения t')
plt.ylabel('Значения f')
plt.show()
```

##### Ограничьте график по оси x значениями 0.5 и 9.5, а по оси y - значениями -2.5 и 2.5.


```python
plt.axis([0.5, 9.5, -2.5, 2.5])
plt.plot(t, f, color='g')
plt.title('График f(t)')
plt.xlabel('Значения t')
plt.ylabel('Значения f')
plt.show()
```

## *Задание 3


##### С помощью функции linspace библиотеки Numpy создайте массив x из 51 числа от -3 до 3 включительно.

```python
x = np.linspace(-3, 3, 51)
x
```

##### Создайте массивы y1, y2, y3, y4 по следующим формулам:
##### y1 = x**2
##### y2 = 2 * x + 0.5
##### y3 = -3 * x - 1.5
##### y4 = sin(x)


```python
y1 = x ** 2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)
y1, y2, y3, y4
```

##### Используя функцию subplots модуля matplotlib.pyplot, создайте объект matplotlib.figure.Figure с названием fig и массив объектов Axes под названием ax,причем так, чтобы у вас было 4 отдельных графика в сетке, состоящей из двух строк и двух столбцов. В каждом графике массив x используется для координат по горизонтали.В левом верхнем графике для координат по вертикали используйте y1,в правом верхнем - y2, в левом нижнем - y3, в правом нижнем - y4.Дайте название графикам: 'График y1', 'График y2' и т.д.


```python
fig, ax = plt.subplots(2, 2)
axes_info = tuple(zip(range(1, 5), ax.flatten(), (y1, y2, y3, y4)))
for i, ax, y in axes_info:
    ax.plot(x, y)
    ax.set_title(f'График y{i}')
plt.subplots_adjust(wspace=0.3, hspace=0.3)
plt.show()
```

##### Для графика в левом верхнем углу установите границы по оси x от -5 до 5. 

```python
fig, ax = plt.subplots(2, 2)
axes_info = tuple(zip(range(1, 5), ax.flatten(), (y1, y2, y3, y4)))
for i, ax, y in axes_info:
    ax.plot(x, y)
    ax.set_title(f'График y{i}')
    if i == 1:
        ax.set_xlim(-5, 5)
plt.subplots_adjust(wspace=0.3, hspace=0.3)
plt.show()
```

##### Установите размеры фигуры 8 дюймов по горизонтали и 6 дюймов по вертикали.

```python
fig, ax = plt.subplots(2, 2)
fig.set_size_inches(8, 6)
axes_info = tuple(zip(range(1, 5), ax.flatten(), (y1, y2, y3, y4)))
for i, ax, y in axes_info:
    ax.plot(x, y)
    ax.set_title(f'График y{i}')
    if i == 1:
        ax.set_xlim(-5, 5)
plt.subplots_adjust(wspace=0.3, hspace=0.3)
plt.show()
```

##### Вертикальные и горизонтальные зазоры между графиками должны составлять 0.3.

##### Импортируйте библиотеку Pandas, а также используйте для графиков стиль “fivethirtyeight”.


```python
import pandas as pd
plt.style.use('fivethirtyeight')
credit_card = pd.read_csv('creditcard.csv')
```

##### Посчитайте с помощью метода value_counts количество наблюдений для каждого значения целевой переменной Class и примените к полученным данным метод plot, чтобы построить столбчатую диаграмму.

```python
classes_quan = credit_card['Class'].value_counts()
classes_quan.plot(kind='barh')
plt.show()
```

##### Затем постройте такую же диаграмму, используя логарифмический масштаб.

```python
classes_quan.plot(kind='barh', logx=True)
plt.show()
```

##### На следующем графике постройте две гистограммы по значениям признака V1 - одну для мошеннических транзакций (Class равен 1) и другую - для обычных (Class равен 0). Подберите значение аргумента density так, чтобы по вертикали графика было расположено не число наблюдений, а плотность распределения. Число бинов должно равняться 20 для обеих гистограмм, а коэффициент alpha сделайте равным 0.5, чтобы гистограммы были полупрозрачными и не загораживали друг друга. Создайте легенду с двумя значениями: “Class 0” и “Class 1”. Гистограмма обычных транзакций должна быть серого цвета, а мошеннических - красного. Горизонтальной оси дайте название “V1”.


```python
class0 = credit_card.loc[credit_card['Class'] == 0, ['V1']]
class1 = credit_card.loc[credit_card['Class'] == 1, ['V1']]
class0['V1'].hist(density=True, bins=20, alpha=0.5, color='grey', label='Class0')
class1['V1'].hist(density=True, bins=20, alpha=0.5, color='red', label='Class1')
plt.xlabel('V1')
plt.legend()
plt.show()
```
